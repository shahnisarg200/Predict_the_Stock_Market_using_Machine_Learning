```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sp500 = yf.Ticker("^GSPC")
```

```python
historical_data = sp500.history(period = "max")
historical_data.head(10)
```

```python
# Extract years and closing prices
years = historical_data.index.year
closing_prices = historical_data['Close']

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(years, closing_prices, color='blue', linewidth=2)

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Closing Price')
plt.title('S&P 500 Closing Prices Over Time')

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
historical_data.drop(columns=['Dividends', 'Stock Splits'], inplace=True)
```

```python
historical_data.head(10)
```

```python
historical_data['Tomorrow'] = historical_data['Close'].shift(-1)
historical_data.head(10)
```

```python
historical_data['Target'] = (historical_data['Tomorrow'] > historical_data['Close']).astype(int)
historical_data
```

```python
historical_data = historical_data.loc['1990-01-01':].copy()
historical_data
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)

train = historical_data.iloc[:-100]
test = historical_data.iloc[-100:]

predictors = ["Close", "Volume", "Open", "High", "Low"]

model.fit(train[predictors], train["Target"])
```

```python
from sklearn.metrics import precision_score

preds = model.predict(test[predictors])
```

```python
preds = pd.Series(preds, index=test.index)
```

```python
precision_score(test["Target"], preds)
```

```python
combined = pd.concat([test["Target"], preds], axis=1)
```

```python
combined.plot()
```

```python
def predict(train, test, predictors, model):
  model.fit(train[predictors], train['Target'])
  preds = model.predict(test[predictors])
  preds = pd.Series(preds, index=test.index, name="Predictions")
  combined = pd.concat([test["Target"], preds], axis=1)
  return combined
```

```python
def backtest(data, model, predictors, start=2500, step=250):
  all_predictions = []
  for i in range(start, data.shape[0], step):
    train = data.iloc[0:i].copy()
    test = data.iloc[i:(i+step)].copy()
    predictions = predict(train, test, predictors, model)
    all_predictions.append(predictions)
  return pd.concat(all_predictions)
```

```python
predictions = backtest(historical_data, model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
horizons = [2,5, 60, 250, 1000]
new_predictors = []

for horizon in horizons:
  rolling_average = historical_data.rolling(horizon).mean()

  ration_column = f"Close_Ratio_{horizon}"
  historical_data[ration_column] = historical_data["Close"] / rolling_average["Close"]

  trend_column = f"Trend_{horizon}"
  historical_data[trend_column] = historical_data.shift(1).rolling(horizon).sum()["Target"]

  new_predictors += [ration_column, trend_column]
```

```python
historical_data = historical_data.dropna()
```

```python
historical_data
```

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
def predict(train, test, predictors, model):
  model.fit(train[predictors], train['Target'])
  preds = model.predict_proba(test[predictors])[:,1]
  preds[preds >= .6] = 1
  preds[preds < .6] = 0
  preds = pd.Series(preds, index=test.index, name="Predictions")
  combined = pd.concat([test["Target"], preds], axis=1)
  return combined
```

```python
predictions = backtest(historical_data, model, new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```
